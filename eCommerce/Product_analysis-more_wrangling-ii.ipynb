#### NOTICED zip_code with four digits and three digits as pandas temoved the leading zeroes. 
#### So padded with zeroes to make it a five-digit format

```python
df_final_customers_clean['zip_code'] = df_final_customers_clean['zip_code'].apply(lambda x: '{:05}'.format(x))
```

```python
df_final_customers_clean.head(15)
```

```python
df_nan_lat_long = df_final_customers_clean[df_final_customers_clean['latitude'].isna()]
```

```python
df_nan_lat_long.head(20)
```

```python
#pip install pgeocode
```

```python
import pgeocode 
nomi = pgeocode.Nominatim('us') 
```

```python
for i in list(df_nan_lat_long.index.values):
    #print(i)
    l_zip = df_nan_lat_long._get_value(i, 'zip_code')
    a=nomi.query_postal_code(l_zip)
    #print(l_zip,a['latitude'], a['longitude'])
    df_nan_lat_long.at[i,'latitude'] = a['latitude']
    df_nan_lat_long.at[i,'longitude'] = a['longitude']
```

```python
df_nan_lat_long.head(20)
```

```python
NaNfreeLat_long_customers = df_final_customers_clean[~df_final_customers_clean['latitude'].isna()]
```

```python
NaNfreeLat_long_customers.shape
```

```python
all_clean_customers_df = pd.concat([NaNfreeLat_long_customers,df_nan_lat_long])
```

```python
all_clean_customers_df.reset_index(inplace=True)
```

```python
all_clean_customers_df.drop(['index','order_city','order_postal','order_state'],axis=1,inplace=True)
```

```python
all_clean_customers_df.head()
```

```python
all_clean_customers_df.info()
```

```python
region_nan_customers = all_clean_customers_df[all_clean_customers_df['region'].isna()]
```

```python
region_less_states = region_nan_customers['state'].unique()
```

```python
region_less_states
```

```python
# Get count of unique values in column pandas
reg_dict={}

for s in region_less_states:
    region_vals= all_clean_customers_df.loc[all_clean_customers_df['state'] == s, 'region'].mode().values
    reg_dict[s] = region_vals[0]
    #print(reg_dict)
```

```python
region_nan_customers['region'] = region_nan_customers['region'].fillna(region_nan_customers['state'].map(reg_dict))
```

```python
region_nan_customers
```

```python
nanFree_reg_customers = all_clean_customers_df[~all_clean_customers_df['region'].isna()]
```

```python
nanFree_reg_customers.shape
```

```python
ultimate_final_customers_df = pd.concat([nanFree_reg_customers,region_nan_customers])
```

```python
ultimate_final_customers_df.shape
```

```python
ultimate_final_customers_df[ultimate_final_customers_df['latitude'].isna()]
```

```python
#fillna with 0.0
ultimate_final_customers_df['latitude'] = ultimate_final_customers_df['latitude'].fillna(0.0)
ultimate_final_customers_df['longitude'] = ultimate_final_customers_df['longitude'].fillna(0.0)
```

```python
ultimate_final_customers_df[ultimate_final_customers_df['latitude'].isna()]
```

```python
ultimate_final_customers_df.info()
```

```python
#saving this clean file
#saving this file
csv_name = "df_ultimate_customers.csv"
ultimate_final_customers_df.to_csv(csv_name, index=False)
```

```python
products = pd.read_csv('products.csv')
```

```python
replace = ['Cleanig Supplies']
final_value = 'Cleaning Supplies'
def replace_values(x):
    if x in replace:
        return final_value
    else:
        return x
products['category'] = products['category'].apply(replace_values)
```

```python
from pathlib import Path
import os
from dotenv import load_dotenv
#load_dotenv()
from sqlalchemy import create_engine
```

```python
if __name__ == '__main__':

    # Loading env variables
    load_dotenv(dotenv_path=Path('.\.env'))

    user = os.getenv('user')
    password = os.getenv('password')
    host = os.getenv('host')
    #port = os.getenv('port')
    database = os.getenv('database')
    print("database: ",database)
```

```python
# Creating Postgre SQL connection, passing user and password as arguments when running script
con_string=f'mysql+pymysql://{user}:{password}@{host}/{database}'
engine = create_engine(con_string)
```

```python
# Exporting to our local instance of MySql
final_sales_clean_df.to_sql('tbl_clean_sales', con=engine, if_exists='replace', index=False)
products.to_sql('tbl_clean_products', con=engine, if_exists='replace', index=False)
ultimate_final_customers_df.to_sql('tbl_clean_customers', con=engine, if_exists='replace', index=False)
```
